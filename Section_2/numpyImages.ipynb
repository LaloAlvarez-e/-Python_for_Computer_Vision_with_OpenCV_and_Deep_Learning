```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from PIL import Image
```

```python
picture = Image.open("../DATA/00-puppy.jpg")
pictureArray = np.asarray(picture)
pictureArray.shape
```

```python
plt.imshow(pictureArray)
plt.axis('off')
plt.show()
```

```python
pictureRedChannel = pictureArray[:,:,0]
plt.imshow(pictureRedChannel, cmap='gray')
plt.axis('off')
plt.show()
```

```python
pictureGreenChannel = pictureArray[:,:,1]
plt.imshow(pictureGreenChannel, cmap='gray')
plt.axis('off')
plt.show()
```

```python
pictureBlueChannel = pictureArray[:,:,2]
plt.imshow(pictureBlueChannel, cmap='gray')
plt.axis('off')
plt.show()
```
